# 蒙特卡洛模拟： 使用Python估算圆周率Pi

### 实验方法
通过随机投点法（Monte Carlo Method)，验证大数定律，并计算Pi的近似值。
### 数学原理
$$\frac{S_{circle}}{S_{square}} = \frac{\pi r^2}{(2r)^2} = \frac{\pi}{4}$$

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成随机点
N = 2000 #生成 N 个
x = np.random.rand(N) #生成2000个0到1之间的随机x坐标
y = np.random.rand(N)

#判断点是否在圆内
#设圆的方程为: x^2 + y^2 <= 1
distance = x**2 + y**2 #来判断有多少点在圆里
inside_circle = distance <= 1 #会生成T/F

M = np.sum(inside_circle) #数有多少点在圆内
pi_estimate = 4 * M / N #计算pi的公式

print(f" {N} ") #展示预设了多少点
print(f"估算的Pi = {pi_estimate}") #输出算出的pi值

plt.figure(figsize=(6, 6)) #构造一个6×6的图表

#x[inside_circle]意思是取判定为T的点
plt.scatter(x[inside_circle], y[inside_circle], color='red', s=10, label='Inside') #scatter代表散点 任务是在坐标系里画图 相对而言plot是连线
plt.scatter(x[~inside_circle], y[~inside_circle], color='blue', s=10, label='Outside')
plt.title(f"Monte Carlo Simulation (Pi) = {pi_estimate})")
plt.legend(loc="upper right") 
plt.show()
```

### 实验结论
当 N 较小的时候，误差较大
随着N逐渐增大 ， 计算结果显著收敛于真实值
验证了**大数定律**
